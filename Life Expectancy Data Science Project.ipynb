## Do densely populated countries tend to have lower life expectancy?

```python
sns.scatterplot(x='Population', y='Life expectancy ', data=df)
```

```python
df['Population'].describe()
```

Since the median population is 1386542, we will use this value to divide the dataframe into countries that are highly
populated (population > 1386542), and countries that are lowly populated (population < 1386542).

```python
def popLevel(x):
    if x <= 1386542:
        x = 'Low'
    else:
        x = 'High'
    return x
```

```python
df['Population Level'] = df['Population'].apply(lambda x: popLevel(x))
df.head(10)
```

```python
high_pop = df[df['Population Level'] == 'High']
print(f"Countries classified as having a high population have an average population of {high_pop['Population'].mean():.2f}.")
low_pop = df[df['Population Level'] == 'Low']
print(f"Countries classified as having a low population have an average population of {low_pop['Population'].mean():.2f}.")
```

```python
df['Life expectancy '].groupby(df['Population Level']).mean()
```

```python
df['Life expectancy '].corr(df['Population'])
```

```python
plt.figure(figsize=(10,5))
plt.title('Life Expectancy vs. Population Level')
sns.barplot(x='Population Level', y='Life expectancy ', data=df)
```

As indicated by the data, there is no relationship between the population of a country and its respective average life
expectancy. Both highly populated and lowly populated countries have an average life expectancy of approximately 69 
years, as indicated by barplot.

To answer the important question, densely populated countries do not tend to have a lower life expectancy.

## Linear Regression

Now that we've explored the data, let's go ahead and split the data into training and testing sets, in order to 
establish a linear model between all other numerical columns and 'life expectancy'. 

```python
non_floats = []
for col in df:
    if df[col].dtypes != "float64" and df[col].dtypes != "int64":
        non_floats.append(col)
df = df.drop(columns=non_floats)
df = df.dropna()
df
```

```python
X = df.drop(columns=['Year', 'Life expectancy '])
y = df['Life expectancy ']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.title('Y Test vs. Predicted Y')
```

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.metrics import r2_score
```

```python
print('The r-squared value for our test data was',r2_score(y_test,predictions))
```

```python
The R2 score can be used as a performance marker of a linear regression model. The r-squared value shows that the
data the model  was fed accounted  for 81.42% of the variance.
```

## Conclusion

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```
